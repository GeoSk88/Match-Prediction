```python
import numpy as np
import matplotlib.pyplot as mtp  
import pandas as pd  

data = pd.read_csv("2011-2019.csv")
data
```

```python
#Feature Selection to Appear Input Dataset in Table 2

drop_cols1 = ['Div', 'Date', 'Referee']
newdf1 = data.drop(columns = drop_cols1)
input_dataset = newdf1.drop(newdf1.loc[:,'B365H':'BbAvAHA'].columns, axis=1)
input_dataset
```

```python
def team_scoring_prediction(input_dataset, home_team, away_team):
    
    total_home_goals = input_dataset['FTHG'].sum() # Το συνολο γκολ που εχουν βαλει ολες οι ομαδες της EPL ως γηπεδουχες
    total_home_goals_mean = (total_home_goals/data.shape[0]).round(6) # Ο μεσος αριθμος γκολ που βαζει μια γηπεδουχος στην EPL
    #print("Average mean Home goal scored in EPL: " +str(total_home_goals_mean))
    
    total_away_goals = input_dataset['FTAG'].sum() # Το συνολο γκολ που εχουν βαλει ολες οι ομαδες της EPL ως φιλοξενουμενες
    total_away_goals_mean = (total_away_goals/data.shape[0]).round(6) # Ο μεσος αριθμος γκολ που βαζει μια φιλοξενουμενη στην EPL
    #print("Average mean Away goal scored in EPL: " +str(total_away_goals_mean))
    
    #Select the home team and away team
    
    home_team = input_dataset.loc[data['HomeTeam'] == home_team]
    away_team = input_dataset.loc[data['AwayTeam'] == away_team]
    
    #Home goals scored by home team and mean
    
    home_team_home_goals_score = home_team['FTHG'].sum() # Το συνολο γκολ που εβαλε η ομαδα ως γηπεδουχος
    home_team_home_goals_score_mean = (home_team_home_goals_score/home_team.shape[0]).round(6) # Ο μεσος αριθμος γκολ που εβαλε η ομαδα ως γηπεδουχος
    #print("Average Home goal scored by Home team in EPL: " + str(home_team_home_goals_score_mean))
    
    #Home goals conceded by home team
    
    home_team_home_goals_concede = home_team['FTAG'].sum() # Το συνολο γκολ που δεχθηκε η ομαδα ως γηπεδουχος
    home_team_home_goals_concede_mean = (home_team_home_goals_concede/home_team.shape[0]).round(6) # Ο μεσος ορος γκολ που δεχθηκε η ομαδα ως γηπεδουχος
    #print("Average Home goal conceded by Home team in EPL: " + str(home_team_home_goals_concede_mean))
    
    #Home team attacking strength 
    home_team_att_strength = (home_team_home_goals_score_mean/ total_home_goals_mean).round(6)
    #print("Home team attacking strength: " + str(home_team_att_strength))
    
    #Home team defensive strength 
    home_team_def_strength = (home_team_home_goals_concede_mean/ total_away_goals_mean).round(6)
    #print("Home team defensive strength: " + str(home_team_def_strength))
    
    #Away goals scored by away team and mean
    
    away_team_away_goals_score = away_team['FTAG'].sum() # Το συνολο γκολ που εβαλε η ομαδα ως φιλοξενουμενη
    away_team_away_goals_score_mean = (away_team_away_goals_score/away_team.shape[0]).round(6) # Ο μεσος αριθμος γκολ που εβαλε η ομαδα ως φιλοξενουμενη
    #print("Average Away goal scored by Away team in EPL: " + str(away_team_away_goals_score_mean))
    
    #Away goals conceded by Away team
    
    away_team_away_goals_concede = away_team['FTHG'].sum() # Το συνολο γκολ που δεχθηκε η ομαδα ως φιλοξενουμενη
    away_team_away_goals_concede_mean = (away_team_away_goals_concede/away_team.shape[0]).round(6) # Ο μεσος αριθμος γκολ δεχνθηκε η ομαδα ως φιλοξενουμενη
    #print("Average Away goal conceded by Away team in EPL: "+ str(away_team_away_goals_concede_mean))
    
    
    #Away team attacking strength 
    away_team_att_strength = (away_team_away_goals_score_mean/ total_away_goals_mean).round(6)
    #print("Away team attacking strength: " + str(away_team_att_strength))
    
    #Away team defensive strength
    away_team_def_strength = (away_team_away_goals_concede_mean/ total_home_goals_mean).round(6)
    #print("Away team defensive strength: " + str(away_team_def_strength))
    
    
    return home_team_att_strength,home_team_def_strength,away_team_att_strength,away_team_def_strength
```

```python
# Feature Engineering
has = []
hds = []
aas = []
ads = []
for index,rdt in data.iterrows():
    home_team = rdt['HomeTeam']
    away_team = rdt['AwayTeam']
    home_team_att_strength, home_team_def_strength, away_team_att_strength, away_team_def_strength = team_scoring_prediction(input_dataset, home_team, away_team)
    has.insert(index,home_team_att_strength) 
    hds.insert(index,home_team_def_strength)
    aas.insert(index,away_team_att_strength)
    ads.insert(index,away_team_def_strength)
input_dataset['HAS'] = has
input_dataset['HDS'] = hds
input_dataset['AAS'] = aas
input_dataset['ADS'] = ads
```

```python
input_dataset
```

```python
#Feature Selection to Appear Dataset in Table3

drop_cols2 = ['FTHG', 'FTAG', 'HTHG', 'HTAG', 'HTR', 'HS', 'AS', 'HF', 'AF', 'HY', 'AY', 'HR', 'AR']
final_dataset = input_dataset.drop(columns = drop_cols2)
final_dataset
```

```python
#Convert Categorical FTR,HomeTeam,AwayTeam to Numeric  [Away = 0, Draw = 1, Home = 2]

final_dataset["HomeTeam_code"] = final_dataset["HomeTeam"].astype("category").cat.codes
final_dataset["AwayTeam_code"] = final_dataset["AwayTeam"].astype("category").cat.codes
final_dataset["FTR_code"] = final_dataset["FTR"].astype("category").cat.codes
final_dataset
```

```python
drop_cols3 = ["HomeTeam", "AwayTeam", "FTR"]
preprocessed_dataset = final_dataset.drop(columns = drop_cols3)
preprocessed_dataset
```

```python
#Extracting Independent and dependent Variable

X = preprocessed_dataset.drop(columns = ['FTR_code'])
y = preprocessed_dataset['FTR_code']
```

```python
# Splitting the dataset into training and test set.  
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.20, random_state=0)  
```

```python
# Feature Scaling  
from sklearn.preprocessing import StandardScaler

scaler_x= StandardScaler()    
x_train= scaler_x.fit_transform(X_train)    
x_test= scaler_x.transform(X_test)  
```

```python
#Using Naive Bayes
from sklearn.naive_bayes import GaussianNB  

model=  GaussianNB() # Create the model
model.fit(x_train, y_train) # Train the model
predictions = model.predict(x_test) # Making Predictions
```

```python
#Calculating accuracy
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, predictions)
print(accuracy)
```

```python
#Creating the Confusion matrix

from sklearn.metrics import confusion_matrix  
cm= confusion_matrix(y_test, predictions)  
cm
```

```python
# Making search space = give values to hyperparameters that you want to search in order to find the best set of values
search_space = {
    "var_smoothing" : np.logspace(0,-9, num=100) #[1e-11, 1e-10, 1e-9]
}
```

```python
from sklearn.model_selection import GridSearchCV
#create the GridSearch object
GS = GridSearchCV(estimator = model,
                  param_grid = search_space,
                  scoring = ['accuracy'],
                  refit = 'accuracy',
                  cv = 5, # cross validation
                  verbose = 4) # how much information we want to print
```

```python
GS.fit(x_train,y_train)
```

```python
print("Best Hyperparameters: ", GS.best_params_) # take the best model with the details that we searched for
print("Best Score: ", GS.best_score_) # take the accuracy of best model
```
