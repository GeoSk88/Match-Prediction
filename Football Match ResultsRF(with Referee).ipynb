```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mtp

data = pd.read_csv("2018-2022(EN).csv")
data
```

```python
def goal_difference(data, half_time_home_goals, half_time_away_goals, full_time_home_goals, full_time_away_goals):
    
    # Calculating Goal Difference in HALF TIME
    if (half_time_home_goals > half_time_away_goals):
        half_time_goal_difference = half_time_home_goals - half_time_away_goals
    elif (half_time_away_goals > half_time_home_goals):
        half_time_goal_difference = half_time_away_goals - half_time_home_goals
    else:
         half_time_goal_difference = half_time_home_goals - half_time_away_goals
            
    # Calculating Goal Difference in FULL TIME
    if (full_time_home_goals > full_time_away_goals):
        full_time_goal_difference = full_time_home_goals - full_time_away_goals
    elif (full_time_away_goals > full_time_home_goals):
        full_time_goal_difference = full_time_away_goals - full_time_home_goals
    else:
         full_time_goal_difference = full_time_home_goals - full_time_away_goals
            
            
    return half_time_goal_difference, full_time_goal_difference
        
```

```python
htgd = []
ftgd = []
for index,row in data.iterrows():
    half_time_home_goals = row['HTHG']
    half_time_away_goals = row['HTAG']
    full_time_home_goals = row['FTHG']
    full_time_away_goals = row['FTAG']
    half_time_goal_difference, full_time_goal_difference = goal_difference(data, half_time_home_goals, half_time_away_goals, full_time_home_goals, full_time_away_goals)
    htgd.insert(index, half_time_goal_difference)
    ftgd.insert(index, full_time_goal_difference)
data['HTGD'] = htgd
data['FTGD'] = ftgd
    
```

```python
#Feature Selection 

drop_cols1 = ['Div', 'Date']
newdf1 = data.drop(columns = drop_cols1)
input_dataset = newdf1.drop(newdf1.loc[:,'B365H':'PSCA'].columns, axis=1)
input_dataset
```

```python
# Convert Categorical Variables to Numeric Variables [Home = 2, Draw = 1, Away = 0]

input_dataset["HomeTeam_code"] = input_dataset["HomeTeam"].astype("category").cat.codes
input_dataset["AwayTeam_code"] = input_dataset["AwayTeam"].astype("category").cat.codes
input_dataset["Referee_code"] = input_dataset["Referee"].astype("category").cat.codes
input_dataset["HTR_code"] = input_dataset["HTR"].astype("category").cat.codes
input_dataset["FTR_code"] = input_dataset["FTR"].astype("category").cat.codes
```

```python
input_dataset
```

```python
drop_cols2 = ["HomeTeam", "AwayTeam", "Referee", "FTR", "HTR"]
preprocessed_dataset = input_dataset.drop(columns = drop_cols2)
preprocessed_dataset
```

```python
#Extracting Independent and dependent Variable

X = preprocessed_dataset.drop(columns = ['FTR_code'])
y = preprocessed_dataset['FTR_code']
```

```python
# Splitting the dataset into training and test set.  
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.30, random_state=0) 
```

```python
# Feature Scaling  
from sklearn.preprocessing import StandardScaler

scaler_x= StandardScaler()    
x_train= scaler_x.fit_transform(X_train)    
x_test= scaler_x.transform(X_test) 
```

```python
#Using Random Forest Algorithm
from sklearn.ensemble import RandomForestClassifier  

model= RandomForestClassifier() # Create the model
model.fit(x_train, y_train) # Train the model
predictions = model.predict(x_test) # Making Predictions
```

```python
#Calculating accuracy
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, predictions)
print(accuracy)
```

```python
#Creating the Confusion matrix for Logistic Regression

from sklearn.metrics import confusion_matrix  
cm= confusion_matrix(y_test, predictions)  
cm
```

```python
# Making search space = give values to hyperparameters that you want to search in order to find the best set of values
search_space = {
    "n_estimators" : [90, 91, 92, 93, 94, 95],
    "max_depth" :[20, 22, 24, 26, 28, 30]
}
```

```python
from sklearn.model_selection import GridSearchCV
#create the GridSearch object
GS = GridSearchCV(estimator = model,
                  param_grid = search_space,
                  scoring = ['accuracy', 'r2'],
                  refit = 'r2',
                  cv = 5,
                  verbose = 4)
```

```python
GS.fit(x_train,y_train)
```

```python
print("Best Hyperparameters: ", GS.best_params_) # take the best model with the details that we searched for
print("Best Scores: ", GS.best_score_) # take the R^2 score of best model
```
